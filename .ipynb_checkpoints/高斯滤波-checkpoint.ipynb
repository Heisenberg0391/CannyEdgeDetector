```python
from __future__ import division
import numpy as np
import cv2
from PIL import Image
import matplotlib.pyplot as plt
```

```python
def gaussian(img):
    # 常用5×5的高斯模板
    kernel = np.array([[2, 4,  5,  2,  2],
               [4, 9,  12, 9,  4],
               [5, 12, 15, 12, 5],
               [4, 9,  12, 9,  4],
               [2, 4,  5,  4,  2]]) / 156

    fil_img = cv2.filter2D(img, -1, kernel)

    return abs(fil_img).astype(int)
```

```python
if __name__ == "__main__":
    img_path = './valve.png'
    im = array(Image.open(img_path))
    im = im[:, :, 0]
    
    plt.gray()

    plt.subplot(1, 2, 1)
    plt.imshow(im)
    plt.axis('off')
    plt.title('Original')

    plt.subplot(1, 2, 2)
    plt.imshow(gaussian(im))
    plt.axis('off')
    plt.title('Filtered')

    plt.show()
```
